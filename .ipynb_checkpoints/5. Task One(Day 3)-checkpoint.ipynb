# DOOR LOCK SYSTEM CHALLENGE – Task One (Day 3)

```python
TRY ONE ! (Open the door first, then Open it again) = DOOR ALREADY OPEN
```

```python
# importing the current date and time
import datetime 

# storing the 'previous date' in a variable
Previous_Date = datetime.datetime.today() - datetime.timedelta()

# assigniing the variable password a value 
password = "password"

# assigning the variables for commands for the door values 
open = "open"
close = "close"
quit = "quit"

# creaing a user input for entering the password
user_input = input("Enter password: ")

# creating a while loop for the incorrect password
while user_input != password:
  print("Error!, Enter your password again")
  user_input = input("Enter password: ")
    
# creating an eles statement for when the password is correctly entered 
else:
    print("Correct password, You have succesfully gotten access to the door")
    
# creating a user input for entering the first command
    user_command = input("Enter a command for the door")
    
#  creating if-elif statements for the first time command
    if user_command == open:
        print("The door is now open")
        print("Door last open at: ", Previous_Date)
    elif user_command == close:
        print("The door is now locked")
        print("Door last looked at: ", Previous_Date)
    elif user_command == quit:
        print("The process has been terminated")
    elif user_command != open or user_command != close or user_command!= quit:
           print("Invalid input")    
            
# creating a user input for entering the second command
    user_command2 = input("Enter another command for the door")
    
# creating if- elif statements for th second time command + Nesting if-else statements inside if-elif statements
    if user_command2 == open:
        if user_command == user_command2:
            print("The door is already open")
        else:    
            print("The door is now open")
            print("Door last open at: ", Previous_Date)
    elif user_command2 == close:
        if user_command == user_command2:
            print("The door is already locked")
        else:
            print("The door is looked")
            print("Door last looked at: ", Previous_Date)
    elif user_command2 == quit:
        print("The process has been terminated")
       
    elif user_command2 != open or user_command2 != close or user_command2 != quit:
            print("Invalid input")
```
